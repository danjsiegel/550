a. N-Gram Frequencies

Using the Wikipedia featured article corpus, find and report the 20 most common bigrams and trigrams. You should include the total number of times the bigram/trigram occurred in the corpus as well as the logarithm of the probability of occurring within the corpus. In mathematical terms, if N is the total number of bigrams or trigrams in the corpus and a particular item occurred m times, then the log frequency is log(m/N). It is common practice to report in logarithmic probabilities when dealing with low probability values.

The following is an example of how to report your findings.

```python
ngrams_to_tag = ''
for section in corpus['section_texts']:
    ngrams_to_tag += section
```

```python
from nltk.util import ngrams
```

```python
bigrams = ngrams(tokenized_ngrams_to_tag,2)

```

```python
trigrams = list(ngrams(tokenized_ngrams_to_tag,3))
counted_trigrams = Counter(trigrams)
df = pd.DataFrame.from_dict(counted_trigrams, orient='index')
df = df.reset_index()
df.columns = ['trigram', 'frequency']
df['Log-Frequency']=  df.apply(lambda row: math.log(row['frequency']/sum(df.frequency)), axis=1)
df.sort_values(by=['frequency'], ascending=False).head(20)
```

```python
def ngram_and_log_freq(tokenized_words, ngram_amt, ngram_type):
    ngram_list = list(ngrams(tokenized_words,ngram_amt))
    counter_grams = Counter(ngram_list)
    data = pd.DataFrame.from_dict(counter_grams, orient='index')
    data = data.reset_index()
    data.columns = [ngram_type, 'frequency']
    data['Log-Frequency']=  data.apply(lambda row: math.log(row['frequency']/sum(data.frequency)), axis=1)
    return data.sort_values(by=['frequency'], ascending=False).head(20)
```

```python
bigrams= ngram_and_log_freq(tokenized_ngrams_to_tag, 2, 'Bigram')
trigrams= ngram_and_log_freq(tokenized_ngrams_to_tag, 3, 'Trigram')
```

# b. N-Gram Frequencies (Removing Stop Words and Punctuation)

Repeat the prior step, but remove stop words and punctuation during the tokenization process. You may need to download stop words from NLTK using the following code.

```python
import string
```

```python
def text_process(mess):
    nopunc = [char for char in mess if char not in string.punctuation and char not in [',','.',';']]
    nopunc = ''.join(nopunc)
    nopunc = nopunc.lower()
    no_stopwords = [word for word in nopunc.split() if word not in stopwords.words('english')]
    nostops =  ' '.join(no_stopwords)
    return nostops.replace("\'", "")
```

```python
no_stops_and_punc = text_process(ngrams_to_tag)
tokenized_no_stops_and_punc = nltk.word_tokenize(str(no_stops_and_punc))
```

```python
no_p_bigrams= ngram_and_log_freq(tokenized_no_stops_and_punc, 2, 'Bigram')
no_p_trigrams= ngram_and_log_freq(tokenized_no_stops_and_punc, 3, 'Trigram')
```

# 3. Entity Extraction

Using chapter 7 of Applied Text Analysis with Python as a guide, implement a function to extract entities from each sentence of the Wikipedia featured articles dataset. You can use an n-gram based approach or a syntax/grammar-based approach. Report on the results of your entity extractor for 20 sentences from articles in the corpus.

```python
from nltk.chunk import conlltags2tree, tree2conlltags
```

```python
to_chunk = ''
for section in corpus['section_texts']:
    to_chunk += section
twenty_sent_to_chunk = nltk.sent_tokenize(to_chunk)[:20]
```

```python
from nltk import ne_chunk
def get_entities(document, labels): 
    entities = []
    tokenized = nltk.word_tokenize(str(document))
    tagged_pos =nltk.pos_tag(tokenized)
    trees = ne_chunk(tagged_pos) 
    iob_tagged = tree2conlltags(trees)
    for i in range(len(iob_tagged)):
        try: 
            if iob_tagged[i][2] in labels:
                entities.append(iob_tagged[i])
        except:
            continue
    return entities
```

```python
get_entities(twenty_sent_to_chunk, ['B-ORGANIZATION', 'B-PERSON'])
```
