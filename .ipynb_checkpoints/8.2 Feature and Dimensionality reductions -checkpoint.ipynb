```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import *
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from scipy.sparse import csr_matrix
from sklearn.decomposition import TruncatedSVD
```

```python
df = pd.read_csv('feature_selection.csv')
```

```python
df.head(5)
```

```python
df.corr().abs()
```

```python
def logreg_on_df(feat, targ, result_type):
    features = StandardScaler().fit_transform(feat)
    X_train, X_test, y_train, y_test = train_test_split(features, targ, test_size=0.33, random_state=42)
    logmodel = LogisticRegression()
    logmodel.fit(X_train,y_train)
    return [result_type, len(feat.columns), logmodel.score(X_test, y_test)]
```

```python
accuracy = []
```

```python
accuracy.append(logreg_on_df(df.drop('label', axis=1), df['label'], 'None'))
```

```python
accuracy
```

Drop correlated Features

```python
corr_matrix = df.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
```

```python
'label' in to_drop
```

```python
print(len(to_drop), len(corr_matrix))
print(len(corr_matrix)- len(to_drop))
```

```python
Noncor = df.copy(deep=True)
```

```python
Noncor = df.drop(to_drop, axis=1)
```

```python
accuracy.append(logreg_on_df(Noncor.drop('label', axis=1), Noncor['label'], 'Drop Correlated Features'))
```

```python
accuracy
```

```python
def TSVD_logreg_on_df(feat, targ, result_type, nfeat):
    features = StandardScaler().fit_transform(feat)
    features_sparse = csr_matrix(features)
    tsvd = TruncatedSVD(n_components=nfeat)
    features_sparse_tsv = tsvd.fit(features_sparse).transform(features_sparse)
    X_train, X_test, y_train, y_test = train_test_split(features_sparse_tsv, targ, test_size=0.33, random_state=42)
    logmodel = LogisticRegression()
    logmodel.fit(X_train,y_train)
    return [result_type, nfeat, logmodel.score(X_test, y_test)]
```

```python
ts = [['TSVD 2', 2],['TSVD 5', 5], ['TSVD 10', 10], ['TSVD 20', 20]]
```

```python
for i in ts:
    accuracy.append(TSVD_logreg_on_df(Noncor.drop('label', axis=1), Noncor['label'], i[0], i[1]))
```

```python
pd.DataFrame(accuracy, columns=['Feat reduction', '# of input Feat', 'Accuracy'])
```

```python
def log_trainer(feat, targ, type_of_reg, no_feats=0):
    X_train, X_test, y_train, y_test = train_test_split(feat, targ, test_size=0.33, random_state=42)
    logmodel = LogisticRegression()
    logmodel.fit(X_train,y_train)
    return [type_of_reg, no_feats, logmodel.score(X_test, y_test)]
```

```python
def Text_TSVD(feat, targ, result_type, nfeat):
    features_sparse = csr_matrix(feat)
    tsvd = TruncatedSVD(n_components=nfeat)
    features_sparse_tsv = tsvd.fit(features_sparse).transform(features_sparse)
    X_train, X_test, y_train, y_test = train_test_split(features_sparse_tsv, targ, test_size=0.33, random_state=42)
    logmodel = LogisticRegression()
    logmodel.fit(X_train,y_train)
    return [result_type, nfeat, logmodel.score(X_test, y_test)]
```

```python
tfidf_vec = TfidfVectorizer()
txt_tf_feat = tfidf_vec.fit_transform(cat_com.txt)
tdd = log_trainer(txt_tf_feat, cat_com['cat'], 'non, tf-idf', no_feats=0)
```

```python
#adding this one manually as it's computationally expensive to run this even though I have the function built
text_acc.append(['None (TF-IDF)', txt_cv_feat.shape[0], tdd[2]])
```
